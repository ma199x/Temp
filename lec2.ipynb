```python
import re
r=re.search("the","The rain  in Spain")
print(r)
```

```python
#ignore character .
import re
r=re.findall("as..","asd#")
print(r)
```

```python
import re
r=re.findall("in","The rain  in Spain")
print(r)
```

```python
#Means that is the first text ^
import re
r=re.findall("^the","The rain in Spain")
print(r)
```

```python
#Means that is the first text \A
import re
r=re.findall("\AThe","The rain in Spain")
print(r)
```

```python
#Means that is the last text $
import re
r=re.findall("^The.*in$","The rain in Spain")
print(r)
```

```python

import re
r=re.findall("in\Z","The rain in Spain")
print(r)
```

```python
#Digit 0-9 \d
import re
r=re.findall("\d\d","The 111 rain 22 in 9 Spain")
print(r)
```

```python
#not Digit 0-9 \D
import re
r=re.findall("\D\D\D","The 1 raa")
print(r)
```

```python
#Digit or letter (small | capital) \w
import re
r=re.findall("\w","a@1 b c A")
print(r)
```

```python
# not Digit or letter (small | capital) \W
import re
r=re.findall("\W\W","a#1@#b(c)")
print(r)
```

```python
import re
r=re.findall("\w\w","ab cd f h t")
print(r)
```

```python
#Space \s
import re
r=re.findall("\s","ab 222 44 55")
print(r)
```

```python
#Or between expression |
import re
r=re.findall("f.r|\w+","for a d INT 98")
print(r)
```

```python
#Or [x or y or n]
import re
r=re.findall("[ab#d]","for #a d INT 98")
print(r)
```

```python
# without [^x and y and n]
import re
r=re.findall("[^abpd]","ptsabewadd")
print(r)
```

```python
#[from-to] [a-z] or [A-Z] or [0-9]
import re
r=re.findall("[a-f]","a d c s e f z h i g k l m 1 2 3 4 5 ")
print(r)
```

```python
import re
r=re.findall("[0-9]","1 2 (6 ,7 , 8 , 9,) 3 4 5")
print(r)
```

```python
import re
r=re.findall("\d\d\d\d\d\d\d\d\d\d\d","my number is : 07811065222 my id is : 1000")
print(r)
```

```python
#repetition {The number of repetitions}
import re
r=re.findall("\d{11}","price : 9999 number ph : 98765432")
print(r)
```

```python
#Included in text + , * , ? , .         \+ , \* , \? , \.
import re
r=re.findall("\?",' what\'s your name ? My name is Ahmad.')
print(r)
```

```python
##Space \S
import re
r=re.findall("from (\S+@\S+)","from asd1@gmail.com to asd2@gmail.com")
print(r)
```
